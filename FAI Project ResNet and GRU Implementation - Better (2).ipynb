```python
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# Define the GRU model
class GRUModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(GRUModel, self).__init__()
        self.gru = nn.GRU(input_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out, _ = self.gru(x)
        out = out[:, -1, :]  # Take the output from the last time step
        return self.fc(out)

# Hyperparameters
input_size = 1000  # The size of your feature vector
hidden_size = 512  # Number of GRU hidden units
output_size = len(set(all_labels.numpy()))  # Number of classes
num_epochs = 50
batch_size = 16
learning_rate = 0.001

# Prepare the dataset and dataloader
dataset = TensorDataset(all_features, all_labels)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Instantiate the model, loss function, and optimizer
model = GRUModel(input_size, hidden_size, output_size)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Training loop
for epoch in range(num_epochs):
    for features_batch, labels_batch in dataloader:
        optimizer.zero_grad()  # Zero gradients
        outputs = model(features_batch)  # Forward pass
        loss = criterion(outputs, labels_batch)  # Calculate loss
        loss.backward()  # Backward pass
        optimizer.step()  # Update weights

    print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')

# Save the model
torch.save(model.state_dict(), 'gru_model.pth')
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

# Define the device based on CUDA availability
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define an enhanced GRU model with more layers and dropout
class EnhancedGRUModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers=3, dropout=0.3):
        super(EnhancedGRUModel, self).__init__()
        
        # Add multiple GRU layers
        self.gru = nn.GRU(input_size, hidden_size, num_layers=num_layers, batch_first=True, dropout=dropout)
        
        # Fully connected layers with increased hidden size
        self.fc1 = nn.Linear(hidden_size, hidden_size * 2)  # Increase the size of the first FC layer
        self.fc2 = nn.Linear(hidden_size * 2, hidden_size)  # Keep the second layer the same size
        self.fc3 = nn.Linear(hidden_size, output_size)  # Final output layer
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(dropout)  # Dropout layer

    def forward(self, x):
        out, _ = self.gru(x)
        out = out[:, -1, :]  # Take the output from the last time step
        out = self.fc1(out)
        out = self.relu(out)
        out = self.dropout(out)  # Apply dropout
        out = self.fc2(out)
        out = self.relu(out)
        out = self.dropout(out)  # Apply dropout again
        return self.fc3(out)

# Hyperparameters
input_size = 1000
hidden_size = 2048  # Increased hidden size for better representation
output_size = len(set(all_labels.numpy()))  # Set output size based on the number of unique labels
num_epochs = 30  # Increased epochs for potentially better training
batch_size = 8
learning_rate = 0.0001

# Prepare the dataset and dataloader
dataset = TensorDataset(all_features, all_labels)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Instantiate the model, loss function, and optimizer
model = EnhancedGRUModel(input_size, hidden_size, output_size).to(device)  # Move model to GPU
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# Training loop
for epoch in range(num_epochs):
    for features_batch, labels_batch in dataloader:
        # Move data to GPU
        features_batch, labels_batch = features_batch.to(device), labels_batch.to(device)

        # Zero gradients
        optimizer.zero_grad()  
        
        # Forward pass
        outputs = model(features_batch)  
        
        # Calculate loss
        loss = criterion(outputs, labels_batch)  
        
        # Backward pass and optimization
        loss.backward()  
        optimizer.step()  

    print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')

# Save the model
torch.save(model.state_dict(), 'enhanced_gru_model.pth')
print("Enhanced Model Saved...!!")
```

```python
import os
import cv2
import numpy as np
import torch
from torchvision import models
from collections import Counter
from tqdm import tqdm  # Import tqdm for the progress bar

# Load the ResNet model
resnet = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
resnet.eval()  # Set to evaluation mode

# Parameters for processing
clip_duration = 20  # Duration of each segment in seconds
frame_rate = 30     # Frame rate of the video (update based on your video)
frames_per_clip = clip_duration * frame_rate  # Total frames per 20-second clip

def extract_features_from_frames(frames):
    features = []
    for frame in frames:
        # Preprocess the frame (resize, normalize, etc.)
        frame = cv2.resize(frame, (224, 224))  # Resize to fit the model
        frame = frame.astype(np.float32) / 255.0  # Normalize

        # Convert to tensor and add a batch dimension
        input_tensor = torch.tensor(frame).permute(2, 0, 1).unsqueeze(0)  # Change to (1, 3, 224, 224)

        with torch.no_grad():
            feature = resnet(input_tensor)  # Forward pass through ResNet
            features.append(feature.view(-1).numpy())  # Flatten and convert to numpy

    # Convert to numpy array and reshape for GRU input
    features = np.array(features)  # Shape will be (num_frames, 2048)
    features = features.reshape(1, features.shape[0], -1)  # Shape will be (1, num_frames, 2048)
    return features

def process_video_in_clips(video_path, model, index_to_label):
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))  # Total frames in the video
    clip_frames = []  # To store frames for the current clip
    predictions = []  # Store predictions for each 20-second clip
    frame_count = 0

    # Use tqdm to show progress over frames
    with tqdm(total=total_frames, desc="Processing Video Frames") as pbar:
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break

            # Accumulate frames for the current clip
            clip_frames.append(frame)
            frame_count += 1
            pbar.update(1)  # Update the progress bar

            # If we've gathered enough frames for a 20-second clip
            if frame_count >= frames_per_clip:
                # Extract features from the accumulated frames
                video_features = extract_features_from_frames(clip_frames)
                video_features_tensor = torch.tensor(video_features, dtype=torch.float32)

                # Make prediction with the GRU model
                with torch.no_grad():
                    outputs = model(video_features_tensor)  # Forward pass
                    _, predicted = torch.max(outputs, 1)  # Get the predicted class

                # Map predicted index to label
                predicted_label = predicted.item()
                predicted_human_label = index_to_label.get(predicted_label, "Unknown Label")
                predictions.append((predicted_human_label, clip_frames.copy()))  # Store label and clip frames

                print(f"Predicted label for 20-second clip: {predicted_human_label}")

                # Reset for next clip
                clip_frames = []
                frame_count = 0

        # Process any remaining frames if they form a partial clip
        if clip_frames:
            video_features = extract_features_from_frames(clip_frames)
            video_features_tensor = torch.tensor(video_features, dtype=torch.float32)

            with torch.no_grad():
                outputs = model(video_features_tensor)
                _, predicted = torch.max(outputs, 1)

            predicted_label = predicted.item()
            predicted_human_label = index_to_label.get(predicted_label, "Unknown Label")
            predictions.append((predicted_human_label, clip_frames.copy()))
            print(f"Predicted label for last partial clip: {predicted_human_label}")

    cap.release()
    return predictions

def merge_clips_with_majority_label(predictions, output_path):
    # Count the occurrences of each label
    labels = [label for label, _ in predictions]
    majority_label = Counter(labels).most_common(1)[0][0]
    print(f"Majority label: {majority_label}")

    # Initialize video writer
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for .mp4 format
    out = None

    # Use tqdm to show progress over merging clips
    with tqdm(total=len(predictions), desc="Merging Clips") as pbar:
        for label, frames in predictions:
            if label == majority_label:
                # Initialize the video writer with the size of the first frame
                if out is None:
                    height, width, _ = frames[0].shape
                    out = cv2.VideoWriter(output_path, fourcc, frame_rate, (width, height))

                # Write frames to output video
                for frame in frames:
                    out.write(frame)

            pbar.update(1)  # Update the progress bar

    if out:
        out.release()
    print(f"Merged video saved at: {output_path}")

# Path to your video file
video_directory = r'D:\FAI_Data_Final'
video_path = os.path.join(video_directory, 'Test_video_LowRes_Small.mp4')  # Update file name as needed

# Assume `model` is your trained GRU model and `index_to_label` is a dictionary mapping indices to labels
predictions = process_video_in_clips(video_path, model, index_to_label)

# Define the output path for the merged video
output_video_path = os.path.join(video_directory, 'Merged_Majority_Label_Video.mp4')

# Merge clips with the majority label and save the output
merge_clips_with_majority_label(predictions, output_video_path)
```
